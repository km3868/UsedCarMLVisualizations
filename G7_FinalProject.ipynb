## EDA:
Data analysis was done with a small random sample of the data, roughly 23%, as the full dataset was too large to work with. As can be seen in the notebook, we preliminarily tagged many features as likely not useful due to:
- high cardinality (`dealer_zip`, `interior_color`)
- duplicate features (`transmission` vs. `transmission_display`, `make` vs `franchise_make`)
- correlated features (`city_fuel_economy` & `highway_fuel_economy`)
- inability to concluslively determine data definition. 

Some features, like torque, are more complex than we feel equipped to deal with, especially without an understanding of how we might be able to aggregate the measure. 
Many of the features require engineering work to strip out unit strings (“39.7 in” for example). 

Looking into the correlations of some of the potential features, many made sense. Higher values of year resulted in higher prices, and higher values for mileage result in lower prices. An interesting observation was that number of days on the market is positively correlated with price, suggesting that cars are sitting when priced too high. Some were surprising though. MPG (both highway and city) are negatively correlated with price, which seems counter-intuitive. 

Because of the high dimensionality of the dataset, additional feature selection will be explored further using either univariate, multivariate, or model-based feature selection techniques.

Next, we will look at the categorical features to get an idea which columns can be onehotencoded or which will need to be dropped due to high cardinality:

```python
#Explore cardinality of categorical features (note: all cols were used but not all are categorical)
dict1={}
for col in data.columns:
  x =len(data[col].value_counts())
  dict1[col]=x
#print(dict1)

sorted_values = sorted(dict1.values(),reverse=True) # Sort the values descending
sorted_dict = {}

for i in sorted_values:
    for k in dict1.keys():
        if dict1[k] == i:
            sorted_dict[k] = dict1[k]
print("column name -- # of unique entries")
for i in sorted_dict:
  print(i,"--",sorted_dict[i])
```

Handling year of car by converting to age buckets

```python
## replace year with age in years?
## age is probably non-linear here. an antique car is old, but valuable, a car from 2000 is likely junk
#first_cut['age'] = 2023 - first_cut.year
```

```python
#categorize age of car into 4 buckets
#antique = 25-45 years old >=1977 and <=1997
#older model = 1997-2010
#used newer = 2011-2015
#newer model = 2015-2022

def yearchange(x):
  if (x <= 1997) & (x >= 1977):
    return(1)  # --> antique
  elif (x>1997) & (x<=2010):
    return(2)  # --> older
  elif (x>2010) & (x<=2015):
    return(3)  # --> newer
  else:
    return(4)  # --> newest
```

```python
first_cut['age'] = first_cut['year'].apply(lambda x: yearchange(x))
```

#### Limitations

Price ranged from 349 to 2,698,500. Subsetting price to reflect "typical" used cars could improve model accuracy at the expense of generalization performance on the entire dataset. Given the dataset size (10gb), only training and testing on a random sample of 730k rows (~23%) could have introduced error and imbalanced data. Features that were dropped could have appeared in the larger dataset and additional features may have been entirely unseen. 

#### Future Work

Because we only utilized a small ~ 700k row sub-sample of the 10gb dataset, it would be interesting to utilize the entire dataset and compare results and performance. Subsetting the dataset on price could also provide a more targeted model for the average used car market and improve performance.

Another expansion topic could be exploring a different feature selection methodology, or allowing the model to do 100% of the feature selection and encoding all of the categorical columns.
